Criar modelos bÃ¡sicos de previsÃ£o

#### Vamos criar a variavel "venue_code" que vai informar se uma equipa jogou em casa ou nÃ£o. 


```python
matches["venue_code"]=matches["venue"].astype("category").cat.codes #Converte a coluna venue para categÃ³rica e depois para cÃ³digos  
matches
```

Vamos criar a varivel opp_code para identificar cada equipa adversÃ¡ria por um nÂº especifÃ­co

```python
matches["opp_code"]= matches["opponent"].astype("category").cat.codes #Converte a coluna opponent para categÃ³rica e depois para cÃ³digos
matches.head()
```

Vamos criar uma variavel que pega apenas a hora da varivel time

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype(int) #Extrai a hora da coluna time e converte para inteiro
matches.head()
```

Vamos criar uma coluna que defina por nÂº os dias da semana: Segunda=0; TerÃ§a=1; Quarta=2;Quinta=3;Sexta=4; SÃ¡bado=5; Domingo=6.

```python
matches["day_code"] = matches["date"].dt.dayofweek #Extrai o dia da semana da coluna date   
matches.head()
```

```python
matches["target"] = (matches["result"]=="w").astype("int") #Cria uma coluna target com valores booleanos
matches.head()
```

## Criar o primeiro modelo de IA

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
#Treino
train = matches[matches["date"]<"2022-01-01"]

#Teste
test = matches[matches["date"]>"2022-01-01"]
```

```python
#Criar a previsÃ£o
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"]) #Treina o modelo
```

```python
preds=rf.predict(test[predictors]) #Faz a previsÃ£o
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(test["target"], preds) #Calcula a acurÃ¡cia
```

```python
acc
```

Combinar os valores actuais e os previstos

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))   #Cria um DataFrame com os valores reais e previstos
```

```python
pd.crosstab(index=combined["actual"], columns=combined["predicted"]) #Cria uma matriz de confusÃ£o
```

```python
from sklearn.metrics import precision_score

```

```python
precision_score(test["target"], preds) #Calcula a precisÃ£o
```

Adicionar a perfomance de cada equipa

```python
grouped_matches= matches.groupby("team") #Agrupa as partidas por time
```

```python
group=grouped_matches.get_group("Manchester City") #Pega as partidas do Manchester City

group.head()
```

```python
def rolling_avarages (group, cols, new_cols): #FunÃ§Ã£o que calcula a mÃ©dia mÃ³vel
    group = group.sort_values("date") #Ordena as partidas por data, crescente
    rolling_states = group[cols].rolling(3, closed="left").mean() #Calcula a mÃ©dia mÃ³vel sobre as 3 Ãºltimas partidas 
    group[new_cols] = rolling_states #Cria uma nova coluna com a mÃ©dia mÃ³vel
    group = group.dropna(subset=new_cols) #Remove os valores nulos
    return group

```

dropna(subset=new_cols): Isso faz com que o DataFrame remova linhas que tenham valores NaN especificamente na(s) coluna(s) new_col (a nova coluna criada com as mÃ©dias mÃ³veis). Ou seja, qualquer linha que nÃ£o tenha uma mÃ©dia mÃ³vel calculada (por exemplo, as primeiras linhas da sÃ©rie, que nÃ£o tÃªm trÃªs partidas anteriores para calcular a mÃ©dia) serÃ¡ removida.

```python
cols = ["gf","ga","sh","dist","fk","pk","pkatt"] #Colunas que serÃ£o usadas para calcular a mÃ©dia mÃ³vel
```

```python
new_cols = [f"{c}_rolling" for c in cols] #Cria uma lista com os nomes das novas colunas
new_cols
```
