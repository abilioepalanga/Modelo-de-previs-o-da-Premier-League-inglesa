Criar modelos básicos de previsão

#### Vamos criar a variavel "venue_code" que vai informar se uma equipa jogou em casa ou não. 


```python
matches["venue_code"]=matches["venue"].astype("category").cat.codes #Converte a coluna venue para categórica e depois para códigos  
matches
```

Vamos criar a varivel opp_code para identificar cada equipa adversária por um nº especifíco

```python
matches["opp_code"]= matches["opponent"].astype("category").cat.codes #Converte a coluna opponent para categórica e depois para códigos
matches.head()
```

Vamos criar uma variavel que pega apenas a hora da varivel time

```python
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype(int) #Extrai a hora da coluna time e converte para inteiro
matches.head()
```

Vamos criar uma coluna que defina por nº os dias da semana: Segunda=0; Terça=1; Quarta=2;Quinta=3;Sexta=4; Sábado=5; Domingo=6.

```python
matches["day_code"] = matches["date"].dt.dayofweek #Extrai o dia da semana da coluna date   
matches.head()
```

```python
matches["target"] = (matches["result"]=="w").astype("int") #Cria uma coluna target com valores booleanos
matches.head()
```

## Criar o primeiro modelo de IA

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
#Treino
train = matches[matches["date"]<"2022-01-01"]

#Teste
test = matches[matches["date"]>"2022-01-01"]
```

```python
#Criar a previsão
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"]) #Treina o modelo
```

```python
preds=rf.predict(test[predictors]) #Faz a previsão
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(test["target"], preds) #Calcula a acurácia
```

```python
acc
```

Combinar os valores actuais e os previstos

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))   #Cria um DataFrame com os valores reais e previstos
```

```python
pd.crosstab(index=combined["actual"], columns=combined["predicted"]) #Cria uma matriz de confusão
```

```python
from sklearn.metrics import precision_score

```

```python
precision_score(test["target"], preds) #Calcula a precisão
```
